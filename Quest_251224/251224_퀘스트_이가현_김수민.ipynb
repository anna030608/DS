1. Account 클래스 생성
- 클래스 데이터 입력 (은행이름, 예금주, 계좌번호, 잔액 설정)
- 계좌번호 3자리-2자리-6자리 랜덤 생성
- 생성된 계좌 count 변수

2. 입금 메서드
- 최소 1원 이상 입금횟수 5회시 1프로 이자 지급 , 입금 내역 기록

3. 출금 메서드
- 계좌 잔고 이상 출금 불가, 출금 내역 기록

4. get_account_num() 메서드
- 생성된 계좌 개수 출력

5. history() 메서드
- 입출금 내역 출력

```python
import random   # 계좌번호 생성을 위해 import

# 클래스 선언
class Account:
  # Account 클래스로 생성된 계좌의 총 개수를 저장하는 변수.
  account_count = 0

  def __init__(self, name, balance):
    # 통장 개설
    self.bank = "모두은행"
    self.name = name

    # 계좌번호 생성 3자리-2자리-6자리 랜덤 번호
    # 000-00-000000
    account_num = f"{random.randint(0 , 999):03d}-{random.randint(0 , 99):02d}-{random.randint(0 , 999999):06d}"
    self.account_num = account_num

    self.balance = balance  # 계좌의 현재 잔액
    self.deposit_count = 0  # 입금 횟수 카운트 (5회마다 이자 지급 확인용 변수)
    self.history_log = []   # 입출금/이자 내역 저장 리스트
    self.count = 0          # 전체 거래 횟수. -> 거래 내역 번호용.

    Account.account_count += 1    # 계좌가 생성될 때마다 전체 계좌 개수 1 증가.


  def deposit(self, amount):  # 입금 메서드
    if amount < 1:            # 1원보다 적은 돈을 입금시 종료
      return False

    # 입금 -> 잔액 증가
    self.balance += amount

    # 입금 룃수 / 전체 거래 횟수 증가.
    self.deposit_count += 1
    self.count += 1
    self.history_log.append(f"{self.count}회: 입금 | 금액: +{amount} | 잔액: {self.balance}")

    # 입금 횟수 5번마다 이자지급
    if self.deposit_count % 5 == 0:
      interest = int(self.balance * 0.01)  # 이자 금액
      self.balance += interest  # 현재 잔액에 이자 더하기.
      self.count += 1   # 이자 지금 -> 전체 거래 횟수 1 증가
      self.history_log.append(f"{self.count}회: 이자지급 | 금액: +{interest} | 잔액: {self.balance}")


  def withdraw(self, amount):  # 출금 메서드
      # 출금 요청 금액이 잔액보다 작거나 같을 경우만, 출금 가능
      if amount <= self.balance:
        self.balance -= amount
        self.count += 1
        self.history_log.append(f"{self.count}회: 출금 | 금액: -{amount} | 잔액: {self.balance}")

      else:  # 잔액 부족 시 안내 메시지 출력.
        print(f"잔액이 부족합니다. (요청: {amount}원 / 잔액: {self.balance}원)")


  def get_account_num(self):    # 현재까지 생성된 전체 계좌 수 반환 메서드.
    return Account.account_count


  def history(self):    # 거래 내역을 순서대로 출력하는 메서드.
    for i in self.history_log:
      print(i)
```

```python
def main():
  account = Account("홍길동", 10000)
  account.deposit(10000)
  account.deposit(10000)
  account.deposit(10000)
  account.deposit(10000)
  account.deposit(10000)
  account.deposit(10000)

  account.withdraw(10000)
  account.withdraw(10000000)

  print()
  print("----------------거래내역-----------------")
  account.history()

  print()
  print("생성된 계좌 수:", account.get_account_num())
```

```python
if __name__ == "__main__":
  main()
```
