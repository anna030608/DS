```python
# 1ë‹¨ê³„ : ë©”ë‰´íŒ ë§Œë“¤ê¸°
n = int(input("ë©”ë‰´íŒì— ì…ë ¥í•  ë©”ë‰´ ìˆ˜: "))
dic = dict()

for _ in range(n):
    potion = input("í¬ì…˜ ì´ë¦„ì„ ì…ë ¥í•˜ì‹œì˜¤: ")
    price = int(input("ê°€ê²©ì„ ì…ë ¥í•˜ì‹œì˜¤(ì •ìˆ˜): "))
    dic[potion] = price


# =========================
# ë©”ë‰´íŒ ì¶œë ¥ (ë²ˆí˜¸, ì´ë¦„, ê°€ê²©)
# =========================
def print_menu(menu_dict):
    print("\n------------- menu -------------")
    keys = list(menu_dict.keys())
    for i in range(len(keys)):
        print(f"{i+1}. {keys[i]} -> {menu_dict[keys[i]]} ê³¨ë“œ")
    print("0. ì£¼ë¬¸ ì¢…ë£Œ")
    print("--------------------------------\n")


print_menu(dic)


# =========================
# 2~5ë‹¨ê³„ : ì£¼ë¬¸ ì‹œìŠ¤í…œ + ë°˜ë³µ + í• ì¸ + ì˜ˆì™¸ì²˜ë¦¬
# =========================
DISCOUNT_THRESHOLD = 10000  # 1ë§Œ ê³¨ë“œ ì´ìƒ
DISCOUNT_RATE = 0.10        # 10%

while True:  # ì†ë‹˜ ë‹¨ìœ„ ë°˜ë³µ
    order_potions = []
    order_counts = []
    total = 0

    print("ì†ë‹˜ì´ ë“¤ì–´ì™”ìŠµë‹ˆë‹¤! ğŸ§™")
    print_menu(dic)

    keys = list(dic.keys())  # ë²ˆí˜¸(1..n) -> í¬ì…˜ ì´ë¦„ ë§¤í•‘ìš©

    while True:  # ì£¼ë¬¸ ë°˜ë³µ
        try:
            menu_num = int(input("ì£¼ë¬¸í•  ë©”ë‰´ ë²ˆí˜¸ë¥¼ ì…ë ¥í•˜ì„¸ìš” (0: ì¢…ë£Œ): "))
        except ValueError:
            print("ìˆ«ìë¡œ ì…ë ¥í•´ì£¼ì„¸ìš”!\n")
            continue

        # ì£¼ë¬¸ ì¢…ë£Œ
        if menu_num == 0:
            break

        # 5ë‹¨ê³„: ì—†ëŠ” ë²ˆí˜¸ ì˜ˆì™¸ ì²˜ë¦¬
        if menu_num < 1 or menu_num > len(keys):
            print("ì—†ëŠ” ë²ˆí˜¸ì…ë‹ˆë‹¤. ë‹¤ì‹œ ì„ íƒí•´ì£¼ì„¸ìš”\n")
            continue

        # ê°œìˆ˜ ì…ë ¥
        try:
            count = int(input("ëª‡ ê°œë¥¼ ì£¼ë¬¸í•˜ì‹œê² ì–´ìš”?: "))
            if count <= 0:
                print("ê°œìˆ˜ëŠ” 1 ì´ìƒì˜ ì •ìˆ˜ì—¬ì•¼ í•©ë‹ˆë‹¤!\n")
                continue
        except ValueError:
            print("ê°œìˆ˜ëŠ” ìˆ«ìë¡œ ì…ë ¥í•´ì£¼ì„¸ìš”!\n")
            continue

        potion_name = keys[menu_num - 1]
        potion_price = dic[potion_name]

        # ì£¼ë¬¸ ë¦¬ìŠ¤íŠ¸ ì—…ë°ì´íŠ¸
        order_potions.append(potion_name)
        order_counts.append(count)

        # ê¸ˆì•¡ ëˆ„ì 
        total += potion_price * count

        # ì£¼ë¬¸ ì •ë³´ í™•ì¸ ì¶œë ¥(2ë‹¨ê³„ ìš”êµ¬)
        print(f"\nâœ… ì£¼ë¬¸ í¬ì…˜ ë¦¬ìŠ¤íŠ¸: {order_potions}")
        print(f"âœ… ì£¼ë¬¸ ê°œìˆ˜ ë¦¬ìŠ¤íŠ¸: {order_counts}")
        print(f"ğŸ’° í˜„ì¬ í•©ê³„: {total} ê³¨ë“œ\n")

    # ì£¼ë¬¸ ì¢…ë£Œ í›„ ì²˜ë¦¬
    if len(order_potions) == 0:
        print("ì£¼ë¬¸ì´ ì—†ìŠµë‹ˆë‹¤. ë‹¤ìŒ ì†ë‹˜ì„ ê¸°ë‹¤ë¦½ë‹ˆë‹¤.\n")
    else:
        # =========================
        # 3ë‹¨ê³„: ì˜ìˆ˜ì¦ ì¶œë ¥
        # =========================
        print("\n========== ğŸ§¾ ì˜ìˆ˜ì¦ ==========")
        print(f"{'í¬ì…˜ ì´ë¦„':<12}{'ê°œìˆ˜':>6}{'ë‹¨ê°€':>10}{'ê¸ˆì•¡':>12}")
        print("-" * 40)

        recalculated_total = 0
        for i in range(len(order_potions)):
            name = order_potions[i]
            cnt = order_counts[i]
            unit = dic[name]
            line = unit * cnt
            recalculated_total += line
            print(f"{name:<12}{cnt:>6}{unit:>10}{line:>12}")

        print("-" * 40)
        print(f"{'í•©ê³„':<12}{'':>6}{'':>10}{recalculated_total:>12}")

        # =========================
        # 4ë‹¨ê³„: í• ì¸ ì´ë²¤íŠ¸
        # =========================
        final_total = recalculated_total
        if recalculated_total >= DISCOUNT_THRESHOLD:
            print("\nì¶•í•˜í•©ë‹ˆë‹¤! 1ë§Œ ê³¨ë“œ ì´ìƒ êµ¬ë§¤ë¡œ 10% í• ì¸ì´ ì ìš©ë˜ì—ˆìŠµë‹ˆë‹¤!")
            final_total = int(recalculated_total * (1 - DISCOUNT_RATE))
            print(f"í• ì¸ ì „ ê¸ˆì•¡: {recalculated_total} ê³¨ë“œ")
            print(f"í• ì¸ í›„ ê¸ˆì•¡: {final_total} ê³¨ë“œ")

        print(f"\nâœ… ìµœì¢… ê²°ì œ ê¸ˆì•¡: {final_total} ê³¨ë“œ")
        print("==============================\n")

    # ë‹¤ìŒ ì†ë‹˜?
    again = input("ë‹¤ìŒ ì†ë‹˜ì„ ë°›ìœ¼ì‹œê² ìŠµë‹ˆê¹Œ? (y/n): ").strip().lower()
    if again != "y":
        print("ê°€ê²Œ ë¬¸ì„ ë‹«ìŠµë‹ˆë‹¤! ğŸ‘‹")
        break
```
